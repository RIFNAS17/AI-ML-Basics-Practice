# AI/ML Glossary (30 Terms)

**Overfitting** – Model learns training data too much and fails on new data.  
**Underfitting** – Model is too simple and misses patterns.  
**Precision** – True positives ÷ predicted positives.  
**Recall** – True positives ÷ actual positives.  
**Accuracy** – Correct predictions ÷ total predictions.  
**F1-Score** – Harmonic mean of precision and recall.  
**Bias** – Error from wrong assumptions in the model.  
**Variance** – Error from sensitivity to training data changes.  
**Supervised Learning** – Training with labeled data (input-output pairs).  
**Unsupervised Learning** – Training with unlabeled data to find patterns.  
**Reinforcement Learning** – Learning via rewards and penalties.  
**Feature Scaling** – Normalizing input features for better training.  
**Gradient Descent** – Optimization algorithm to minimize loss.  
**Loss Function** – Measures difference between predicted and actual values.  
**Cross-Validation** – Technique to evaluate model performance on unseen data.  
**Training Set** – Data used to train a model.  
**Test Set** – Data used to evaluate model performance.  
**Validation Set** – Data used to tune hyperparameters.  
**Hyperparameters** – Settings chosen before training (e.g., learning rate).  
**Epoch** – One full pass through the training dataset.  
**Batch Size** – Number of samples processed before updating weights.  
**Learning Rate** – Step size in gradient descent.  
**Confusion Matrix** – Table showing true vs predicted classifications.  
**ROC Curve** – Graph showing trade-off between true positive and false positive rates.  
**AUC (Area Under Curve)** – Measure of model’s ability to distinguish classes.  
**Clustering** – Grouping similar data points without labels.  
**Dimensionality Reduction** – Reducing number of features (e.g., PCA).  
**Neural Network** – Model inspired by human brain with layers of nodes.  
**Activation Function** – Function that introduces non-linearity (e.g., ReLU, Sigmoid).  
**Regularization** – Technique to reduce overfitting (e.g., L1, L2).
