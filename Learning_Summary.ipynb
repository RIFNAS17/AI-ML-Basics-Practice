# Learning Summary

## Supervised Learning
- Uses labeled data (input-output pairs).
- Example: Predicting house prices.
- Common algorithms: Linear Regression, Decision Trees.

## Unsupervised Learning
- Uses unlabeled data to find hidden patterns.
- Example: Customer segmentation (clustering).
- Common algorithms: K-Means, PCA.

## Reinforcement Learning
- Agent learns by interacting with environment.
- Rewards for good actions, penalties for bad ones.
- Example: Training a robot to walk.

## Key Takeaways
- Supervised = labeled data.
- Unsupervised = unlabeled data.
- Reinforcement = trial and error with rewards.
